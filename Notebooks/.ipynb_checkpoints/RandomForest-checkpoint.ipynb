# RandomForestRegression

```python
# Importamos las paqueterías
import sklearn
import pandas as pd
from sklearn import ensemble, datasets
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
ubi = r'C:\Users\Frank\Documents\GitHub\AI-project-2025-1\data\resultados_eventos.csv'
dataset = pd.read_csv(ubi)
dataset
```

```python
# Datos a considerar:
# 1 Attempt
# 3 Foul
# 8 Free kick won
# 11 Penalty conceded

char = []
labels = []
for i in range(len(dataset)):
    #aux = [int(dataset['event_type1'][i]), int(dataset['event_type3'][i]), int(dataset['event_type8'][i]), int(dataset['event_type11'][i])]
    aux = [int(dataset['event_type1'][i]), int(dataset['event_type3'][i]), int(dataset['event_type4'][i]), int(dataset['event_type9'][i]), int(dataset['event_type11'][i])]
    char.append(aux)
    labels.append(int(dataset['event_type2'][i]))
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    char, 
    labels,
    test_size = 0.3,
    random_state = 420
)
```

```python
# Declaración de modelo y entrenamiento
regresor = RandomForestRegressor(max_depth = 3, random_state = 3)
regresor.fit(xtrain, ytrain)
```

```python
# Predicción
test_pred = regresor.predict(xtest)
```

```python
# Función de evaluación
from scipy import stats
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def eval(test_labels, test_predictions):
    spear = stats.spearmanr(test_labels, test_predictions)
    r2 = r2_score(test_labels, test_predictions)
    mse = mean_squared_error(test_labels, test_predictions)
    mae = mean_absolute_error(test_labels, test_predictions)
    print(f"MAE: {mae}, MSE: {mse}, R2: {r2}, Spearman: {spear.statistic}")
```

```python
# Resultados
eval(ytest, test_pred)
```
